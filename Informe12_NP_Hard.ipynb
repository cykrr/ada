#1. Preguntas

##1.1. ¿Qué quiere decir que un problema sea NP-Hard?

Que un problema sea NP-Hard implica que pertenece a una clase de problemas que son reconocidos informalmente como aquellos "que son al menos igual de difíciles que los problemas más complicados NP". 

De modo formal, se puede plantear que los problemas se categorizan como NP-Hard si contiene los problemas H tales que todo problema en $L$ en NP puede ser reducido polinomialmente en $H$. Esto es asumiendo que la solución de $H$ toma $1$ unidad de tiempo, con lo que la solución de H podría ser utilizada para resolver $L$ en tiempo polinomial. 

##1.2 ¿Qué quiere decir que $P \neq NP$?

Cabe decir que esto es una conjetura, por lo cual no está probado. Esto quiere decir que no existen algoritmos que puedan resolver problemas **NP** en tiempo polinomial. 

Para ser mas claros:

**Clasificación P**: Son aquellos problemas "tratables", los cuales se pueden resolver en tiempo polinomial.

**Clasificación NP**: Son aquellos problemas verificables en tiempo polinomial, 

A partir de esto podemos darnos cuenta que no se puede afirmar que **P = NP**, ya que los problemas NP no se pueden resolver en tiempo polinomal, es por esto que se toma la conjetura $P \neq NP$.

##1.3  ¿Cómo se prueba que un problema es NP-Hard usando reducciones?

Para demostrar que un problema $B$ es **NP-Hard** se debe seleccionar un problema $A$, el cual se sabe que es **NP-Hard** y si el problema $A$ se puede reducir al problema $B$ se demuestra que $B$ es **NP-Hard**.

De tal manera que si:

$A \in NP-Hard$ $\rightarrow$ $B \in NP-Hard$

## SubsetSum

El problema de la suma de subconjuntos **SSP** es un problema de decisión. Este consiste en: A partir de un conjunto de $S$ números enteros y una suma **objetivo** $T$, verificar si existe algún subconjunto tal que la suma de los números es $T$.

Por ejemplo:

Dado el subconjunto $S$: $[1,3,4,5,8]$ y $T = 17$

La salida sería **True**, ya que existe el subconjunto $[4, 5, 8]$, tal que la suma de este es $4$.


Entrada: Una serie de numeros positivos $a_1, a_2,..., a_n$ y un numero entero positivo $T$ el cual es la suma a buscar.

Salida: **True** en caso de encontrar un subconjunto, tal que la suma sea $T$ o **False** de no encontrarlo.

Teorema: El **Conjunto Independiente máximo** se reduce a la **Suma de subconjuntos**.

Corolario: El problema de la suma de subconjuntos es NP-Hard

##Reducción de Conjunto Independiente máximo a Suma de subconjuntos

Definimos una instancia del Conjunto Independiente máximo (IND), el problema del conjunto independiente, como un string de la forma $\langle G ⟩ \langle k ⟩$ donde $G$ es un grafo y $k$ un entero positivo. Ese string es un miembro del lenguaje IND si es que hay algún conjunto de $I$ vértices de $G$ de modo que:

$|I|$ = $k$ y no hay dos números de $I$ que forman un arco en $G$, un conjunto **k-independiente** de $G$.

Definimos una instancia de SubsetSum, el problema de la suma de subconjuntos como un string $S$ $T$ donde $S$ es una lista de números positivos y $T$ es un número. Decimos que $\langle S ⟩ \langle T ⟩$ ∈ SubsetSum si existe algun subconjunto de $S$, tal que su suma sea $T$. Probamos que **SSP** es NP, usando el método de certificado. La sublista que su total es igual a $k$ es el certificado, el cual puede ser (trivialmente) verificado en tiempo polinomial. 

Ahora definimos una reducción P-Time R de IND para Subset_sum, donde IND es el problema
del conjunto independiente. Asumimos que todos nuestros lenguajes (problemas) están 
sobre un alfabeto $∑$. Sin perder la generalidad, $∑$ es el alfabeto binario. Cada reducción debe ser una función:

R: $∑^*$ $→$ $∑^*$

Cuando definimos $R(w)$ abajo, vamos a asumir que $w$ es una instancia del **IND**. Si w es cualquier otro string, definimos R(w) = λ, el string vacío.

Dejemos  $\langle G ⟩ \langle k ⟩$ como una instancia de IND, donde G = (V,E). Escribimos $V = {v_1, v_2...v_n}$, los vértices de G, y $E = {e_1, e_2...e_n}$ los arcos de G. Decimos que $e_j$ se encuentra con $v_i$, y escribe $e_j \perp v_i$, si $v_i$ es uno de los puntos finales de $e_j$.

Ahora definimos R($\langle G ⟩ \langle k ⟩$) = $\langle S ⟩ \langle T ⟩$ una instancia del problema de la suma de subconjuntos. 
Se define $weight(v_i)$ = $10^{m+1} + ∑_{j:e_j\perp v_i } 10^j$ y $weight(e_j) = 10^j$. Dejamos $S$ = $weight(v_1)...weight(v_n), weight(e_i)...weight(e_n)$, y dejamos $T = k ⋅ 10^{m+1} + ∑_{j = i}^m 10^j$

Si seguimos los dos siguientes lemas, R es una reducción de IND para Subset_sum.

**Lema 1**: Si $G$ tiene un conjunto independiente de tamaño $k$, entonces $\langle S ⟩ \langle T ⟩$ ∈ **SubsetSum**.

Prueba:

Dejamos $I$ como un conjunto de $k$ vértices independientes de $G$. Fijamos $J$ como el conjunto de arcos que no se encuentra con ninguno de los vértices en $I$. Fijamos $X$ como la suma de los pesos de los vértices en $I$ y los arcos en $J$, eso es, $X$ es la suma de la subsecuencia $S$.
Escribimos $S$ = $\sum_| l = 1^{m+1}a_l  10^l$.

Veredicto: $S = K$.

Prueba del veredicto: 

Como $I$ tiene una cardinalidad $k$, tenemos $l_m+1 = k$ ya que $I$ tiene cardinalidad $k$. Para cualquier $1<=l<=m$, si $e_l$ no se encuentra con ningún miembro de $I$, $y_e$ contribuye $10^l$ a $X$, mientras que si $e_l$ se encuentra con $v_i$, entonces $x_i$ contribuye $10^l$ a $X$. Ya que $I$ es **independiente**, $e_l$ no se encuentra con ningún otro vértice, entonces no hay contribución adicional de $10^l$ a $X$. Eso es, $α_l$ = 1
en cualquiera de los dos casos. De este modo, $S = K$. 

**Lema 2**: Si$\langle S ⟩ \langle T ⟩$ ∈ subsetSum entonces $$ tiene un conjunto independiente de tamaño $k$.

Pruebas: Suponga que $T$ es la suma de un subconjunto de $S$. Entonces existen conjuntos $I$ ⊆ $V$ y $J$ ⊆ E de modo que $K$ es la suma de los pesos de un $I$
⊆ $V$ y $J$ ⊆ $V$. Escribimos $K$ = $\sum_| l = 1^{m+1}a_l  10^l$. Ya que $α_{m+1}$ = $k$, la cardinalidad de $I$ es $k$. No hay dos miembros de $J$ que abarquen un borde, ya que de otro modo $α_l$ = 2 para algunos $l$. De este modo, $I$ es un conjunto **k-independiente** de $G$.

#4. Descripción de algoritmos

## SubsetSum

Una solución para este problema es recorrer todos los subconjuntos de $n$ números y comprobar si alguno de estos suma el número $T$. Esta solución tendría un tiempo de ejecución de $O(2^n \cdot n)$, esto se debe a que existen un total de $2^n$ subconjuntos, y para cada subconjunto debemos sumar como máximo $n$ números.

Este algoritmo se puede implementar utilizando la recursión, en donde existen dos posibilidades:

Caso base: No quedan elementos en el arreglo.

1. Se inclye el elemento actual en el subconjunto y se repite para los **n-1** elementos restantes.

2. Se excluye el elemento actual del subconjunto y se repite para los **n-1** elementos restantes.

Cabe decir que este algoritmo posee una complejidad espacial de $O(n)$.






Por otro lado existe otro algoritmo el cual posee un menor tiempo de ejecución de $O(2^{n/2}\cdot n)$, pero requiere un espacio mucho mayor al mencionado anteriormente $O(2^{n/2})$. Este algoritmo divide de manera arbitraria el subconjunto en dos, de tamaño $n/2$ cada uno. Para cada conjunto se calcula la suma de todos los posibles subconjuntos $O(2^{n/2})$. Estos dos vectores se ordenan en un tiempo de $O(2^{n/2} \cdot n)$. A partir de esto se puede verificar si un elemento del primer arreglo más un elemento del segundo dan el total $T$ buscado en tiempo $O(2^{n/2})$. Para lograr esto es necesario que el algoritmo pase por el primer arreglo del orden establecido (de forma decreciente) y después pasando por el más pequeño (orden creciente). Si la suma del elemento en turno de los dos arreglos es el doble que S el algoritmo procede a moverse al siguiente elemento en el primer arreglo, y cuando es menor a S se mueve al siguiente elemento en el segundo arreglo. Cuando se encuentra S el algoritmo termina su ejecución.

### 3PAT

Asuma que se tienen dos contenedores. Podemos llenar ambos al mismo tiempo
y considerar las sobras en un tercer contenedor.


Para llenar los contenedores podemos ayudarnos del algoritmo DP (programación dinámica)
del problema de la mochila 0-1. Cuya complejidad es $O(wn)$ siendo w el peso
máximo entre de la mochila y n la cantidad de elementos. Esto nos permite
desarrollar una función recursiva para solucionar el problema 3PAT, por lo
que la solución a utilizar será también DP. Note como el problema 3PAT es una variación tri-dimensional del bi-dimensional problema de la mochila.

Se define $P[x,y,i]$ como un subproblema donde el elemento $i$ se puede introducir
o en un contenedor de tamaño $x$, o en un contenedor de tamaño $y$.

Inicialización: Sabemos que el problema $P[0, 0, 0]$ es correcto puesto que
no hay ningún elemento en el contenedor y no se inserta nada. Para el resto de casos
base $P[x,y,0]$ sabemos que no cumplen con el requerimiento ya que la dimensión
de las contenedor será distintas de cero y no estamos insertando nada.

Para el caso general $P[x,y,i]$ se dan tres posibles casos:

* El elemento no cabe en ninguno de los dos contenedores. Se toma la iteración anterior
$P[x,y,i-1]$

* El elemento cabe en el contenedor de tamaño $x$, se usa la iteración del contenedor
$x$ anterior $P[x-a_i, y, i - a_i]$

* El elemento cabe en el contenedor de tamaño $y$, se usa la iteración de el contenedor $x$ anterior $P[x, y-a_i, i - a_i]$.

Si cualquiera de estas condiciones se cumple entonces $P[x,y,i]$ también.

Esto tiene una complejidad $O(Tn)$. Siendo $T$ la suma máxima del conjunto inicial y $n$ la cantidad de elementos.

Uno podría intuir incorrectamente que esto tiene una complejidad polinomial. Sin embargo a medida que aumenta $T$ la cantidad de memoria requerida para solucionar el problema aumenta exponencialmente. Esto es debido a que en binario la cantidad de numero representables con una determinada cantidad de bits es menor y es más probable que se produzcan overflows al operar dichos valores. Requiriendo asi una mayor cantidad de bits para almacenar los valores
obtenidos con la funcion de recursión.

### TETRIS

Para resolver un problema de tetris se propone el siguiente algoritmo luego de una correcta inicialización del tablero (en tiempo polinomial).


```
para $i=0 ... i < a_i$:

    Inicialización: Se inserta una pieza "L" en el  
    contenedor actual

    Se insertan las piezas "O", "J", "O"

    Finalización: Se insertan las piezas "O" y "I" para     
    "reiniciar" y poder insertar un nuevo elemento

si el contenedor está lleno: insertar "L" para cerrar

si s contenedores están llenos:
  insertar pieza "T" en el bloqueo para abrirlo
  llenar el área de relleno con 5T + 16 bloques
```
Si el conjunto de entrada cumple con el requisito del problema. Este se vaciará, de lo contrario se tendrán piezas remanentes en el tablero.
